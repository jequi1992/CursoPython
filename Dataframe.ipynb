```python
import seaborn as sns
from seaborn import load_dataset
tips=load_dataset("tips")
tips
```

```python
from seaborn import boxplot
p=boxplot(df.edad, orient="h", color="red",fliersize = 8, linewidth=2)
p.set(xlabel="edad")
```

```python
from seaborn import boxplot
p=boxplot(df.fallas, orient="h", color="black",fliersize = 9, linewidth=5)
p.set(xlabel="Numero Fallas")
```

```python
df=df[df["edad"]<=100]
df=df[df["fallas"]<=100]
df
```

```python
import matplotlib.pyplot as plt
columnum=["Estudiante","edad","sexo","municipio","nota prog","fallas"]
fig,ax=plt.subplots(nrows=6,ncols=1,figsize=(10,30))
fig.subplots_adjust(hspace=1)
for i,col in enumerate(columnum):
    sns.countplot(x=col, data=df, ax=ax[i])
    ax[i].set_title(col)
    ax[i].set_xticklabels(ax[i].get_xticklabels())
```
